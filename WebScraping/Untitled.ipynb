```python
from requests import get
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time
from IPython.display import clear_output  
import pandas as pd
import csv
     
```

```python
pages =[str(i) for i in range(1,39)]
```

```python
# 1. Redéclarer les listes
nom = []
adresse = []
telephone =[]
        
# 2. préparer l'affichage des boucles
start_time = time()
requests = 0
              
# 4. boucle sur starts contenant les premiers films des pages de 1 à 4
for page in pages:
# 5. Faire une requête GET
    response = get('https://www.goafricaonline.com/bj/annuaire/restaurants?p='+ page)
# 6. Pause la boucle de 8 à 15 secondes
    sleep(randint(8,15))
            
# 7. Afficher les informations sur les requêtes
    requests += 1
    elapsed_time = time() - start_time
    print('Request: {}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)
    
# 8. Avertir si le code status est différent de 200
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))
                
# 9. Stopper la boucle si le nombre de requêtes est supérieur à 38
    if requests > 38:
        warn('Nombre de requêtes trop important')
        break
# 10. Extraire le HTML avec BeautifulSoup
    page_html = BeautifulSoup(response.text, 'html.parser')
            
# 11. Sélectionner les 50 films de chaque page (container)
    mv_containers = page_html.find_all('div',class_="flex w-full")
            
# 12. Boucle pour chaque container
    for container in mv_containers:
        if container.find(class_='flex flex-1 t:ml-10 min-w-180px') is not None:
# 14. scrape le nom
            name = container.a.text
            nom.append(name)
                    
#15. scrape l'adresse
            address = container.address.text
            adresse.append(address)
                    
#16. scrape le téléphone
            imdb = container.find(class_="flex flex-1 t:ml-10 min-w-180px").a.text
            telephone.append(imdb)
    
```

```python
name =[]
for i in nom :
    t =i.replace('\n','')
    name.append(t)

address =[]
for i in adresse :
    t =i.replace('\n','')
    address.append(t)

phone =[]
for i in telephone :
    t =i.replace('\n','')
    phone.append(t)
    
```

```python
tableau = pd.DataFrame({
    'nom' : name,
    'adresse': address,
    'telephone': phone
    })
```
